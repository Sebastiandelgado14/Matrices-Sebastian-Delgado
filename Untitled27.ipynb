```python
usuarios = {
    "ANA": "1234",
    "JUAN": "abcd"
}

try:
    usuario = input("Usuario: ").upper()
    password = input("Contraseña: ")

    if usuario not in usuarios:
        raise KeyError("El usuario no existe")

    if usuarios[usuario] != password:
        raise ValueError("Contraseña incorrecta")

    else:
        print("Bienvenido,", usuario)

except KeyError as e:
    print("Error:", e)

except ValueError as e:
    print("Error:", e)
```

El error ocurre porque los valores ingresados con input() son almacenados por Python como texto (str). Aunque el usuario escriba números, el programa los interpreta como cadenas de caracteres y no como valores numéricos. Por esta razón, al intentar realizar una operación matemática como la división, Python no puede operar entre textos Esto significa que el operador / no puede utilizarse entre dos variables de tipo str. Para solucionar el problema, es necesario convertir los datos ingresados a números utilizando float() o int(). De esta manera, Python reconocerá los valores como numéricos y podrá realizar las operaciones matemáticas correctamente.



```python
try:
    datos = input("Ingresa números separados por coma: ")
    lista = [int(x) for x in datos.split(",")]

    indice = int(input("Índice a consultar: "))
    print("Elemento:", lista[indice])

except ValueError:
    print("Debes ingresar solo números")

except IndexError:
    print("El índice está fuera de rango")

except TypeError:
    print("Debes ingresar números enteros mayores que 1")
```

```python

productos = {
    "pan": 1.5,
    "leche": 2.0,
    "huevos": 3.0
}

print("Productos disponibles:", productos)

try:
    nombre = input("¿Qué producto quieres?: ").lower()

    # Verificar primero si el producto existe
    if nombre not in productos:
        raise KeyError("El producto no está disponible")

    # Solo pide cantidad si el producto sí existe
    cantidad = int(input("Cantidad: "))

    if cantidad <= 0:
        raise ValueError("La cantidad debe ser mayor a 0")

    total = productos[nombre] * cantidad

    print("Total a pagar:", total)

except ValueError as e:
    print("Error:", e)

except KeyError as e:
    print("Error:", e)

except Exception as e:
    print("Ocurrió un error inesperado:", e)
```

```python
try:

    datos = input("Ingresa notas separadas por coma: ")

    # Convertir cada valor a float correctamente
    notas = [float(n) for n in datos.split(",")]

    if len(notas) == 0:
        raise ValueError("No ingresaste notas")

    promedio = sum(notas) / len(notas)

    posicion = int(input("¿Qué nota quieres ver (índice)?: "))

    if posicion < 0:
        raise ValueError("Ingresa solo índices positivos")

    print("Nota seleccionada:", notas[posicion])

    print("Promedio:", promedio)

except ValueError as e:
    print("Error de valor:", e)

except ZeroDivisionError:
    print("No se puede dividir entre cero")

except IndexError:
    print("Índice fuera de rango")
```
